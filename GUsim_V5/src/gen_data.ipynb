```python
import json
import random
import copy

keys = ['formality', 'conciseness', 'sentiment', 'carelessness', 'emojis', 'impatient', 'vague']
formality = ['1', 'f1']
conciseness = ['1', 'f1', '0']
sentiment = ['1', 'f1', '0']
proactiveness = ['1', 'f1']
carelessness = ['1']
impatient = ['1']
vague = ['1']
emojis = ['1']

key_to_values = {
    'formality': formality,
    'conciseness': conciseness,
    'sentiment': sentiment,
    'carelessness': carelessness,
    'emojis': emojis,
    'impatient': impatient,
    'vague': vague
}


# 设置随机种子

import time
# 进行 5 次采样
for i in range(42, 46):
    random.seed(time.time())
    # 每次深拷贝原始数据
    datas_with_behavior = []
    
    # 随机选择 2 或 3 个属性

    for data in datas:
        selected_keys = random.sample(keys, k=random.choice([2, 3]))
        new_data = copy.deepcopy(data)
        new_data['user_behavior'] = {}

        # 对每个 selected_key 随机赋值
        for key in selected_keys:
            new_data['user_behavior'][key] = random.choice(key_to_values[key])
        
        datas_with_behavior.append(new_data)

    # 写入文件
    with open('profiles_w_f.jsonl', 'a', encoding='utf-8') as file:
        for data in datas_with_behavior:
            file.write(json.dumps(data, ensure_ascii=False) + '\n')
```
